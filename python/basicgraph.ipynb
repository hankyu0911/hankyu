```python
mu,sigma  = 100,15
x = mu+sigma*np.random.randn(10000)
plt.hist(x,bins=50,density=True,facecolor='g',alpha=0.5)
plt.title('IQ histogram')
plt.xlabel('Smart')
plt.ylabel('Probability')
plt.text(60,.025,r'$\mu=100, \ \sigma=15$')
plt.axis([40,160,0,0.03])
plt.grid(True)
plt.show()
```

```python
t = np.arange(0.,5.,0.2)

plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.legend(['같은값','제곱','세제곱'])
plt.show()
```

```python
line1 = plt.plot(t,t,label='같은값')
line2 = plt.plot(t,t**2,label='제곱')
line3 = plt.plot(t,t**3,label='세제곱')

plt.setp(line1,color='red',linestyle='--',marker='o')
plt.setp(line2,color='blue',linestyle='-.',marker='s')
plt.setp(line3,color='green',linestyle=':',marker='^')

plt.legend(loc='upper right')
plt.show()
```

```python
def f_graph(t):
    return np.exp(-t)* np.cos(2*np.pi*t)

x = np.arange(0.,5.,0.1)
fig,axes = plt.subplots(2,sharex=True)

fig.suptitle('Sharing x axis')
axes[0].plot(x,f_graph(x),'bo')
axes[1].plot(x,np.cos(2*np.pi*x),'r--')
plt.show()
```

```python
x = np.linspace(0,2 * np.pi,400)
y = np.sin(x **2)

f,axes = plt.subplots(2,2,sharex='col', sharey = 'row', figsize=(10,6))
f.suptitle('Sharing x per column, y per row')
axes[0,0].plot(x,y)
axes[0,1].scatter(x,y)
axes[1,0].scatter(x,y,color='red')
axes[1,1].plot(x,y,color='red')
plt.show()
```
